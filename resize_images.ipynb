```
import cv2
import os
#import numpy as np
```

```
## Resize

os.chdir('/content/drive/MyDrive/MangaColoring/Berserk/labeled/bw/bw')
for path in os.listdir():
    image = cv2.imread(path)
    image = cv2.resize(image, (256, 256))
    path = path.split('.')[0] + '_256x256.png'
    cv2.imwrite(path, image)
```

```
## Concatenate for pix2pix

os.chdir('/content/drive/MyDrive/MangaColoring/Berserk/labeled/bw/bw')
for path in os.listdir():
    image_bw = cv2.imread(path)
    image_bw.squeeze()

    os.chdir('/content/drive/MyDrive/MangaColoring/Berserk/labeled/rgb/rgb')
    image_rgb = cv2.imread(path)

    image = np.zeros(shape=(256, 512, 3), dtype=np.float)

    image[:256, :256] = image_bw
    image[:256, 256:] = image_rgb

    path = path.split('.')[0] + '.png'

    os.chdir('/content/drive/MyDrive/MangaColoring/Berserk/labeled/combined_bw_rgb')
    cv2.imwrite(path, image)

    os.chdir('/content/drive/MyDrive/MangaColoring/Berserk/labeled/bw/bw')
    
```
